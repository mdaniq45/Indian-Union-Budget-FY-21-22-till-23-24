## Indian Union Budget FY 21-22 till 23-24

#### The Indian Economy is currently the 5th Largest Economy in the world. And it aims to reach among the Top 3 in the next 3-4 years. It has registered tremendous economic growth over the last decade.

The 2023 Union Budget of India was presented by the Minister of Finance of India on February 01, 2023.

The Union Budget for FY 2023-24 aims to further strengthen India's economic status. In the 75th Year of India's Independence, the World has recognized the Indian Economy as a 'bright star' with its Economic Growth estimated at 7 per cent, which is the highest among all major economies.

The Vision for** 'Amrit Kaal'** articulated in the Union Budget for FY 2023-24 is centered around:

Opportunities for Citizens with focus on youth
Growth & Job creation
Strong & Stable Macro-Economic Environment
The seven priorities, termed Saptarishi, adopted in the Union Budget for FY 2023-24 to guide the country towards 'Amrit Kaal', thus providing a blueprint for an empowered and inclusive economy, are:

Inclusive Development
Reaching the last mile
Infrastructure & Investment
Unleashing the potential
Green Growth
Youth Power
Financial Sector
The dataset provides details about India's allocation of budget to different schemes in various Ministries under Government of India since Fiscal Year 2021-22 to FY 23-24.

The exploratory analysis of the dataset will provide important views about where the Government is targeting in the last few years so as to bring about such drastic change in the economy.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df['Budget Estimates 2022-2023 Revenue'].fillna(df['Budget Estimates 2022-2023 Revenue'].mean(), inplace=True)
```

```python
df['Budget Estimates 2022-2023 Capital'].fillna(df['Budget Estimates 2022-2023 Capital'].mean(), inplace=True)
```

```python
# Strip any leading/trailing spaces in column names
df.columns = df.columns.str.strip()
```

```python
# Check for missing values
print(df.isnull().sum())

# Get basic statistics
print(df.describe())

```

```python
# Selecting relevant columns
df_bar = df[['Category', 'Actuals 2021-2022 Revenue', 'Budget Estimates 2022-2023 Revenue', 
             'Revised Estimates 2022-2023 Revenue', 'Budget Estimates2023-2024 Revenue']]

# Setting the 'Category' column as index
df_bar.set_index('Category', inplace=True)

# Plot
df_bar.plot(kind='bar', figsize=(14, 8), stacked=False)
plt.title('Budget Estimates, Actuals, and Revised Estimates (Revenue)')
plt.xlabel('Categories')
plt.ylabel('Amount (in crores)')
plt.xticks(rotation=90)
plt.legend(title='Estimate Types')
plt.tight_layout()
plt.show()
```

```python
# Plotting line plots for Revenue trends
plt.figure(figsize=(14, 8))

sns.lineplot(data=df[['Actuals 2021-2022 Revenue', 'Budget Estimates 2022-2023 Revenue', 
                      'Revised Estimates 2022-2023 Revenue', 'Budget Estimates2023-2024 Revenue']],
             dashes=False, marker="o")

plt.title('Revenue Trends: Actuals, Budget Estimates, and Revised Estimates')
plt.xlabel('Index')
plt.ylabel('Revenue (in crores)')
plt.legend(title='Revenue Types')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(14, 8))
sns.boxplot(data=df[['Actuals 2021-2022 Revenue', 'Actuals 2021-2022 Capital', 
                     'Actuals 2021-2022 Total', 'Budget Estimates 2022-2023 Revenue', 
                     'Budget Estimates 2022-2023 Capital', 'Budget Estimates 2022-2023 Total']])
plt.title('Box Plot of Financial Data to Detect Outliers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title('Correlation Heatmap of Financial Data')
plt.tight_layout()
plt.show()
```

```python
df.rename(columns={'Revised Estimates 2022-2023 Total': 'Revised Estimates2022-2023 Total'}, inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Budget Estimates 2022-2023 Total', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.title("Budget Allocation Per Category (2022-23)")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_grouped = df.groupby("Ministry/Department")["Budget Estimates 2022-2023 Total"].sum()
df_grouped.nlargest(10).plot(kind="pie", autopct="%1.1f%%", cmap="coolwarm")
plt.ylabel("")  # Hide the y-label
plt.title("Top 10 Ministries by Budget Share (2022-23)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_grouped = df.groupby("Category")[["Actuals 2021-2022 Total", "Budget Estimates 2022-2023 Total", "Budget Estimates2023-2024 Total"]].sum()
df_grouped.plot(kind="bar", figsize=(14, 6))
plt.title("Budget Allocation by Category (Comparison Across Years)")
plt.xticks(rotation=90)
plt.ylabel("Budget Amount")
plt.legend(title="Year")
plt.show()
```
